```python
# Membuat fungsi sin(x) kita sendiri
import matplotlib.pyplot as plt
def sin(x):
    x = x % (2 * 3.14159)  # Membuat domain x, misal nilai x adalah = 3.14159 maka sisa pembagian adalah 0, nilai sisa pembagian 0 menjadi domain awal (0, 2pi) dan nilai 2pi didapat dari 4*3.14159 / 2*3.14159 = 2*pi
    
    if x > 3.14159: #nilai x akan dioperasikan ketika melebihi 3.14159 atau pi
        x = x - 3.14159
    term = x
    sum = x
    for i in range(1, 80):
        term *=  -1*x**2 / ((2 * i) * (2 * i + 1)) # Membuat deret Mclauren, dengan mengalikan dengan satu suku sebelumnya
        sum += term
    return sum

# Membuat list untuk sumbu x dan y
x = [0.1 * i for i in range(-314, 315)]
y = [sin(i) for i in x]

plt.plot(x,y)
plt.show()


```
